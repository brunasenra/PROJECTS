```python
import urllib.request
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from numpy.random import randn
np.random.seed(101)
```

```python
wiki = 'https://en.wikipedia.org/wiki/List_of_NGC_objects_(1%E2%80%931000)#1%E2%80%93100'
page = urllib.request.urlopen(wiki)
```

```python
soup = BeautifulSoup(page, 'html5lib')
```

```python
all_table = soup.find_all('table')
```

```python
table = soup.find('table', class_='wikitable sortable')
print (table)
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]

for row in table.findAll("tr"): #para tudo que estiver em <tr>
    cells = row.findAll('td') #variável para encontrar <td>
    if len(cells)==7: #número de colunas
        A.append(cells[0].find(text=True)) #iterando sobre cada linha
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
        G.append(cells[6].find(text=True))
```

```python
df = pd.DataFrame(index=A, columns=['Nomenclatura NGC'])

df['Nomenclatura NGC']=A
df['Outros Nomes']=B
df['Tipos de Objetos']=C
df['Constelação']=D
df['Ascenção Reta']=E
df['Declinação']=F
df['Magnitude Aparente']=G

df
```
